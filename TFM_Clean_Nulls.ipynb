# Abrimos el csv generado para quitar los NaN, ordenar y modificar valores y columnas

```python
import pandas as pd
import csv

df3 = pd.read_csv('df_null.csv')
```

```python
df3.head()
```

```python
df3.isnull().sum().sum()
```

```python
df3.notnull().all()
```

### Columna CO_DETECTOR_PRESENT_DESC 

```python
df3["CO_DETECTOR_PRESENT_DESC"] = df3["CO_DETECTOR_PRESENT_DESC"].fillna(value=0)
```

```python
df3["CO_DETECTOR_PRESENT_DESC"].head()
```

#### Cambio "Yes" por 1 y "NaN" por 0

```python
df3["CO_DETECTOR_PRESENT_DESC"] = df3["CO_DETECTOR_PRESENT_DESC"].replace({"Yes":"1"}, regex=True)
```

```python
df3.head()
```

### Columna UNITS_ONSCENE 

#### Sustituyo los NaN con la media de esa columna

```python
df3["UNITS_ONSCENE"] = df3["UNITS_ONSCENE"].fillna(df3["UNITS_ONSCENE"].mean())
```

```python
df3["UNITS_ONSCENE"].isnull().sum()
```

```python
df3["UNITS_ONSCENE"].head()
```

```python
df3["UNITS_ONSCENE"].astype(int)
```

### Columna ZIP_CODE

```python
df3["ZIP_CODE"].isnull().sum().sum()
```

```python
df3["ZIP_CODE"].isnull().any(axis=0)
```

```python
#Ordenamos el dataframe para poner los valores nulos al principio

zip_code_orden = df3.sort_values("ZIP_CODE")
zip_code_orden.tail(10)
```

#### Al ordenarlo me doy cuenta que cuando no se sabe el ZIP_CODE en el dataframe esta puesto 99999, as√≠ que he decidido sustituir los NaN por 99999.

```python
df3["ZIP_CODE"] = df3["ZIP_CODE"].fillna(value=99999)
df3.tail()
```

```python
df3["ZIP_CODE"].isnull().sum().sum()
```

### Columna ARRIVAL_DATE_TIME                   

```python
df3["ARRIVAL_DATE_TIME"].isnull().sum().sum()
```

#### Sustituyo los NaN que tengo en esta columna con los datos de la columna anterior, ya que la diferencia de tiempo entre que llega el aviso y acuden los bomberos es peque√±a. 

```python
df3['ARRIVAL_DATE_TIME'] = df3['ARRIVAL_DATE_TIME'].fillna(df3['INCIDENT_DATE_TIME'])
```

```python
df3['ARRIVAL_DATE_TIME'].isnull().sum().sum()
```

### Columna LAST_UNIT_CLEARED_DATE_TIME 

```python
df3["LAST_UNIT_CLEARED_DATE_TIME"].isnull().sum().sum()
```

```python
# Ordeno el dataframe por esta columna para localizar los NaN

last_unit_orden = df3.sort_values("LAST_UNIT_CLEARED_DATE_TIME")
last_unit_orden.tail()
```

#### Sustituyo los dos valores NaN con sus correspondientes en la columna INCIDENT_DATA_TIME

```python
df3['LAST_UNIT_CLEARED_DATE_TIME'] = df3['LAST_UNIT_CLEARED_DATE_TIME'].fillna(df3['INCIDENT_DATE_TIME'])

#df3["LAST_UNIT_CLEARED_DATE_TIME"][616721] = df3["LAST_UNIT_CLEARED_DATE_TIME"].fillna(value="12/03/2017 11:19:00")
```

```python
df3["LAST_UNIT_CLEARED_DATE_TIME"].isnull().sum().sum()
```

### Columna HIGHEST_LEVEL_DESC  

```python
df3["HIGHEST_LEVEL_DESC"].isnull().sum().sum()
```

#### Sustituyo los NaN con el dato v√°lido anterior

```python
df3["HIGHEST_LEVEL_DESC"] = df3["HIGHEST_LEVEL_DESC"].fillna(method = "ffill")
```

```python
df3["HIGHEST_LEVEL_DESC"].isnull().sum().sum()
```

### Columna TOTAL_INCIDENT_DURATION 

```python
df3["TOTAL_INCIDENT_DURATION"].isnull().sum().sum()
```

#### Sustituyo los NaN con la media de los valores de esa columna

```python
df3["TOTAL_INCIDENT_DURATION"] = df3["TOTAL_INCIDENT_DURATION"].fillna(df3["TOTAL_INCIDENT_DURATION"].mean())
```

```python
df3["TOTAL_INCIDENT_DURATION"].isnull().sum()
```

### Columna PROPERTY_USE_DESC

```python
df3["PROPERTY_USE_DESC"].isnull().sum().sum()
```

```python
property_use_orden = df3.sort_values("PROPERTY_USE_DESC")
property_use_orden.tail(8)
```

#### Sustituyo los NaN con el dato v√°lido anterior

```python
df3["PROPERTY_USE_DESC"] = df3["PROPERTY_USE_DESC"].fillna(method = "ffill")
```

```python
df3["PROPERTY_USE_DESC"].isnull().sum().sum()
```

### Columna STREET_HIGHWAY  

```python
df3["STREET_HIGHWAY"].isnull().sum().sum()
```

#### Finalmente decido eliminar esta columna ya que la direcci√≥n de la calle la tengo de forma m√°s completa en la columna Adress

```python
df3 = df3.drop(['STREET_HIGHWAY'], axis=1)
```

```python
df3.head()
```

```python
df3.isnull().sum().sum()
```

```python
df3.notnull().all()
```

#### Divido las columnas "Property_use_desc" e "Incident_type_desc" en dos para separar el c√≥digo de la descripci√≥n

```python
df4 = df3.join(df3["PROPERTY_USE_DESC"].str.split('-', 1, expand=True).rename(columns={0:"CODE_PROPERTY", 1:"PROPERTY_TYPE"}))
```

```python
df5 = df4.join(df3["INCIDENT_TYPE_DESC"].str.split('-', 1, expand=True).rename(columns={0:"CODE_INCIDENT", 1:"INCIDENT_TYPE"}))
```

```python
df5["CO_DETECTOR_PRESENT_DESC"].unique()
```

```python
df5["CO_DETECTOR_PRESENT_DESC"] = df5["CO_DETECTOR_PRESENT_DESC"].replace({"No":"2"}, regex=True)
```

```python
var_number = df5["UNITS_ONSCENE"].astype(int)

df5["UNITS_ONSCENE"] = var_number.replace({2.07425684:2}, regex=True)
```

```python
df5["UNITS_ONSCENE"].unique()
```

```python
df5["ZIP_CODE"] = df5["ZIP_CODE"].astype(int)
```

```python
df5["BOROUGH_DESC"] = df5["BOROUGH_DESC"].replace({"X":"BRONX",
                                                 "B":"BROOKLYN",
                                                 "M":"MANHATTAN",
                                                 "Q":"QUEENS",
                                                 "R": "STATEN ISLAND"}, regex=True)
```

```python
df5.columns
```

```python
df5.shape
```

```python
df5.notnull().all()
```

#### Las columnas en las que tengo fechas las pongo como datetime

```python
df5["INCIDENT_DATE_TIME"] = pd.to_datetime(df5["INCIDENT_DATE_TIME"])
```

```python
df5["ARRIVAL_DATE_TIME"] = pd.to_datetime(df5["ARRIVAL_DATE_TIME"])
```

```python
df5["LAST_UNIT_CLEARED_DATE_TIME"] = pd.to_datetime(df5["LAST_UNIT_CLEARED_DATE_TIME"])
```

```python
df5["TOTAL_INCIDENT_DURATION"] = df5["TOTAL_INCIDENT_DURATION"] / 60  #se pasa a minutos
```

#### Utilizo la columna Incident_date_time como √≠ndice del dataframe y lo ordeno

```python
df6 = df5.set_index("INCIDENT_DATE_TIME")

df6.head()
```

```python
df7 = df6.sort_values(["INCIDENT_DATE_TIME"])
```

```python
df7.head()
```

```python
df8 = pd.DataFrame(data=df7)
```

```python
df8.head()
```

#### Guardo el csv resultante

```python
df8.to_csv('df_clean.csv', header=True, index=True)
```
